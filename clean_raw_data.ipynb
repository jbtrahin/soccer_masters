```python
# get one position per player 
clean_player_df['Pos'] = clean_player_df['Pos'].map(lambda x: x.split(',')[0])
clean_player_df
```

```python
# Remove player instances with no minutes played
clean_player_df = clean_player_df[clean_player_df['Min'].notna()]
clean_player_df['Player'].value_counts()
```

```python
# Remove all none match stats - players
clean_player_df = clean_player_df.drop(['#', 'Nation'], axis=1)
clean_player_df['Player'].value_counts()
```

```python
def position_group (row):
    if row['Pos'] == 'FW' or row['Pos'] == 'RW' or row['Pos'] == 'LW':    
        return 'Forward'
    if row['Pos'] == 'CM' or row['Pos'] == 'DM' or row['Pos'] == 'AM' or row['Pos'] == 'MF':    
        return 'Midfield'
    if row['Pos'] == 'CB' or row['Pos'] == 'RB' or row['Pos'] == 'LB' or row['Pos'] == 'WB' or row['Pos'] == 'DF':    
        return 'Defender'
    return row['Pos']
```

```python
clean_player_df['position_group'] = clean_player_df.apply (lambda row: position_group(row), axis=1)
```

```python
clean_player_df
```

```python
clean_player_df['position_group'].value_counts()
```
