```python
from tqdm import tqdm
import requests
from splinter import Browser
import pandas as pd
from bs4 import BeautifulSoup as bs
import matplotlib as plt
from sklearn.datasets import make_blobs
import sklearn as skl
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from tensorflow import keras
```

```python
def extract_link_from_td(td):
    children = list(td.children)
    if len(children)>0:
        return  'https://fbref.com' + children[0].attrs['href']
    else:
        return None
    
url_matches_2019 = 'https://fbref.com/en/comps/9/schedule/Premier-League-Fixtures'
url_matches_2018 = 'https://fbref.com/en/comps/9/1889/schedule/2018-2019-Premier-League-Fixtures'
url_matches_2017 = 'https://fbref.com/en/comps/9/1631/schedule/2017-2018-Premier-League-Fixtures'
urls_matches = [
    url_matches_2019,
    url_matches_2018,
    url_matches_2017
]

all_match_report_link_hrefs = []

for url in urls_matches:
    
    response = requests.get(url)
    soup = bs(response.text)
    match_report_link_td_selector = 'td[data-stat="match_report"]'
    match_report_link_tds = soup.select(match_report_link_td_selector)
    match_report_link_hrefs = list(filter(lambda href: href, map(extract_link_from_td, match_report_link_tds)))
    #print(match_report_link_hrefs)
    all_match_report_link_hrefs.append(match_report_link_hrefs)
    #print(all_match_report_link_hrefs)
```

```python
# Testing
#pd.read_html(all_match_report_link_hrefs[0])[3]

player_df = pd.DataFrame({})
goalie_df = pd.DataFrame({})

for total_matches in tqdm(all_match_report_link_hrefs):
    print(total_matches)
    
    for links in tqdm(total_matches):
        player_df = pd.concat([player_df, pd.read_html(links)[3]])
        goalie_df = pd.concat([goalie_df, pd.read_html(links)[4]])
        player_df = pd.concat([player_df, pd.read_html(links)[5]])
        goalie_df = pd.concat([goalie_df, pd.read_html(links)[6]])
        # print(home_df)
        # print(home_g_df)
        # print(away_df)
        # print(away_g_df)
```
