```python
forward_data = player_data[player_data['Position Group'] == 'Forward']
forward_data
```

```python
# Groupby player (forward) mean
forward_data_mean = forward_data.groupby(["Player"]).mean()
forward_data_mean.head()
```

```python
def backwardElimination(x, sl):
    numVars = len(x[0])
    for i in range(0, numVars):
        regressor_OLS = sm.OLS(y_var, x).fit()
        maxVar = max(regressor_OLS.pvalues).astype(float)
        if maxVar > sl:
            for j in range(0, numVars - i):
                if (regressor_OLS.pvalues[j].astype(float) == maxVar):
                    x = np.delete(x, j, 1)
    regressor_OLS.summary()
    return x
```

```python
# Create target
y_var = forward_data_mean.Gls
y_var.head()
```

```python
player_x = forward_data_mean.drop("Gls", axis =1)
player_x.head()
```

```python
# Create features 
X = player_x
X.head()
```

```python
X_opt = X.iloc[:,:].values
X_opt
```

```python
y_var = y_var.reset_index().iloc[:,1:].values
```

```python
# Test if features are siginiicant 
trimmed_X = backwardElimination(X_opt, 0.05)
trimmed_X
```

```python
# Split our preprocessed data into our features and target arrays
y = y_var
X = trimmed_X
```

```python
# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
# Create scaler instance
scaler = StandardScaler()
```

```python
# Fit the scaler
X_scaler = scaler.fit(X_train)
```

```python
# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Define the linear regression model
lin_classifier = LinearRegression()
```

```python
# Train the model
lin_classifier.fit(X_train_scaled, y_train)
```

```python
lin_classifier.score(X_train, y_train)
```

```python
lin_classifier.score(X_test, y_test)
```

```python
# Define hidden layers
number_input_features = len(X_train_scaled[0])
hidden_layer1 = 50
hidden_layer2 = 50
```

```python
nn = tf.keras.models.Sequential()
```

```python
# Hidden layer 1
nn.add(tf.keras.layers.Dense(units=hidden_layer1, input_dim=number_input_features, activation="relu"))
```

```python
# Hidden layer 2
nn.add(tf.keras.layers.Dense(units=hidden_layer2, input_dim=number_input_features, activation="relu"))
```

```python
# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Check the structure of the model
nn.summary()
```

```python
# Train the model
fit_model = nn.fit(X_train_scaled, y_train, epochs=100)
```

```python
# Evaluate the model using the test data - mean
model_loss, model_accuracy = nn.evaluate(X_test_scaled, y_test, verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
