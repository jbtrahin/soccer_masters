```python
import pandas as pd
import numpy as np
import sklearn as skl
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score
from tensorflow import keras
from sklearn.metrics import r2_score
import statsmodels.api as sm
```

```python
player_data = pd.read_csv('/Users/hunter/Desktop/Data_Class/final_project/Data/cleaned_players.csv')
```

```python
player_data
```

```python
forward_data = player_data[player_data['Position Group'] == 'Forward']
forward_data
```

```python
# Groupby player (forward) mean
forward_data_mean = forward_data.groupby(["Player"]).mean()
forward_data_mean.head()
```

```python
forward_players = forward_data_mean.filter(['Player'], axis=1)
forward_players
```

```python
def backwardElimination(x, sl):
    numVars = len(x[0])
    for i in range(0, numVars):
        regressor_OLS = sm.OLS(y_var, x).fit()
        maxVar = max(regressor_OLS.pvalues).astype(float)
        if maxVar > sl:
            for j in range(0, numVars - i):
                if (regressor_OLS.pvalues[j].astype(float) == maxVar):
                    x = np.delete(x, j, 1)
    regressor_OLS.summary()
    return x
```

```python
# Create target
y_var = forward_data_mean.SoT
y_var.head()
```

```python
player_x = forward_data_mean.drop("SoT", axis =1)
player_x.head()
```

```python
# Create features 
X = player_x
X.head()
```

```python
X_opt = X.iloc[:,:].values
X_opt
```

```python
y_var = y_var.reset_index().iloc[:,1:].values
```

```python
# Test if features are siginiicant 
trimmed_X = backwardElimination(X_opt, 0.05)
trimmed_X
```

```python
# Split our preprocessed data into our features and target arrays
y = y_var
X = trimmed_X
```

```python
# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
# Create scaler instance
scaler = StandardScaler()
```

```python
# Fit the scaler
X_scaler = scaler.fit(X_train)
```

```python
# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Define the linear regression model
lin_reg = LinearRegression()
```

```python
# Train the model
lin_reg.fit(X_train_scaled, y_train)
```

```python
lin_reg.score(X_train, y_train)
```

```python
lin_reg.score(X_test, y_test)
```

```python
y_pred = lin_reg.predict(X_test)
```

```python
y_pred_all = lin_reg.predict(X)
```

```python
r2_score(y_test, y_pred)
```

```python
# Define hidden layers
number_input_features = len(X_train_scaled[0])
hidden_layer1 = 50
hidden_layer2 = 50
```

```python
nn = tf.keras.models.Sequential()
```

```python
# Hidden layer 1
nn.add(tf.keras.layers.Dense(units=hidden_layer1, input_dim=number_input_features, kernel_initializer='normal', activation="relu"))
```

```python
# Hidden layer 2
nn.add(tf.keras.layers.Dense(units=hidden_layer2, input_dim=number_input_features, kernel_initializer='normal', activation="relu"))
```

```python
# Output layer
nn.add(tf.keras.layers.Dense(units=1))
```

```python
# Compile the model
nn.compile(loss="mean_squared_error", optimizer="adam")#, metrics=["accuracy"])
```

```python
# Check the structure of the model
nn.summary()
```

```python
# Train the model
fit_model = nn.fit(X_train_scaled, y_train, epochs=100)
```

```python
# Evaluate the model using the test data - mean
#model_loss, model_accuracy = nn.evaluate(X_test_scaled, y_test, verbose=2)
#print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
model_loss = nn.evaluate(X_test_scaled, y_test, verbose=2)
print(f"Loss: {model_loss}")
```
