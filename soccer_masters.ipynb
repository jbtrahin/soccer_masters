```python
#Import all dependencies
import json
import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
player_stats =[]

for i in range(1059900, 1059920):
    r = requests.get(f'https://api.monpetitgazon.com/championship/match/{i}')
    try:
        results = r.json()
        
        match_id = results['id']
        match_date = results['dateMatch']
        print(f'{i}: {match_id}/{match_date}')
        
        players = results['Home']['players']
        
        for key, value in players.items():
            player_stats.append({
                "player_id": key,
                "player_stat": value['stat'],
                "player_info": value['info']            
            })
            print(f'{i}: success')

    except:
        print(f'{i}: no data')
    continue
    
    
    
    
    
    
    
    
    
    
```
