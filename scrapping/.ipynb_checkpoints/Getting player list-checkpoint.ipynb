```python
#Import all dependencies
import json
import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Extract player list from JSON
player_list =[]

record_count = 1

results = requests.get(f'https://api.monpetitgazon.com/stats/championship/2/2020').json()

for result in range(len(results)):
    print(f"Processing Record {record_count} | {result}")
    record_count += 1  
    
    # Parse out the needed data.
    player_id = results[result]["id"]
    player_lastname = results[result]["lastname"]
    player_firstname = results[result]["firstname"]
    player_position = results[result]["position"]
    player_club = results[result]["club"]
    player_teamid = results[result]["teamId"]
    
    # Append the stats to player_stats.
    player_list.append({
        "player_id": player_id,
        "last_name": player_lastname,
        "first_name": player_firstname,
        "position": player_position,
        "club": player_club,
        "team_id": player_teamid
    })
    
    print(f"{player_lastname} is appended.")
    
    
    
    
```

```python
len(player_list)
```

```python
# Convert the array of dictionaries to a Pandas DataFrame.
player_list_df = pd.DataFrame(player_list)
pd.set_option('display.max_rows',None)
player_list_df
```

```python
#Transform player_id column
player_list_df["player_id"] = player_list_df["player_id"].replace({'player_':''}, regex=True)
player_list_df["player_id"] = player_list_df["player_id"].astype(str).astype(int)
player_list_df
```

```python
#Save as csv
player_list_df.to_csv(index=False)
```
