```python
#Import all dependencies
import json
import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Extract player list from JSON
player_stats =[]

record_count = 1

results = requests.get(f'https://api.monpetitgazon.com/stats/player/95715?season=2017').json()

player_id = results["id"]
player_lastname = results["lastname"]
calendar = results["calendar"]
number_of_matches = results["nbMatch"]
sub_out = results["stats"]["appearances"]["standBy"]
sub_in = results["stats"]["appearances"]["standIn"]
starter = results["stats"]["appearances"]["starter"]
total_appearances = results["stats"]["appearances"]["total"]

matches = results["stats"]["matches"]

for match in range(len(matches)):
    print(f"Processing Record {record_count} | {match}")
    record_count += 1 
    
    match_date = matches[match]["date"], 
    season_day = matches[match]["day"],
    match_id = matches[match]["id"],
    goals = matches[match]["info"]["goals"]
    minutes_played = matches[match]["info"]["minsPlayed"]
    match_rate = matches[match]["info"]["rate"]
    sub = matches[match]["info"]["sub"]
    accurate_back_zone_pass = matches[match]["stats"]["accurate_back_zone_pass"]
    accurate_cross_nocorner = matches[match]["stats"]["accurate_cross_nocorner"]
    accurate_fwd_zone_pass = matches[match]["stats"]["accurate_fwd_zone_pass"]
    accurate_long_balls = matches[match]["stats"]["accurate_long_balls"]
    accurate_pass = matches[match]["stats"]["accurate_pass"]
    att_pen_goal = matches[match]["stats"]["att_pen_goal"]
    att_pen_miss = matches[match]["stats"]["att_pen_miss"]
    att_pen_post = matches[match]["stats"]["att_pen_post"]
    att_pen_target = matches[match]["stats"]["att_pen_target"]
    big_chance_created = matches[match]["stats"]["big_chance_created"]
    big_chance_missed = matches[match]["stats"]["big_chance_missed"]
    clean_sheet = matches[match]["stats"]["clean_sheet"]
    duel_lost = matches[match]["stats"]["duel_lost"]
    duel_won = matches[match]["stats"]["duel_won"]
    error_lead_to_goal = matches[match]["stats"]["error_lead_to_goal"]
    failed_cross = matches[match]["stats"]["failed_cross"]
    fouls = matches[match]["stats"]["fouls"]
    goal_assist = matches[match]["stats"]["goal_assist"]
    goals = matches[match]["stats"]["goals"]
    goals_conceded = matches[match]["stats"]["goals_conceded"]
    interception_won = matches[match]["stats"]["interception_won"]
    own_goals = matches[match]["stats"]["own_goals"]
    percentage_pass = matches[match]["stats"]["percentage_pass"]
    poss_lost_ctrl = matches[match]["stats"]["poss_lost_ctrl"]
    red_card = matches[match]["stats"]["red_card"]
    shot_off_target = matches[match]["stats"]["shot_off_target"]
    total_back_zone_pass = matches[match]["stats"]["total_back_zone_pass"]
    total_contest = matches[match]["stats"]["total_contest"]
    lost_contest = matches[match]["stats"]["lost_contest"]
    total_cross_nocorner = matches[match]["stats"]["total_cross_nocorner"]
    total_fwd_zone_pass = matches[match]["stats"]["total_fwd_zone_pass"]
    total_long_balls = matches[match]["stats"]["total_long_balls"]
    total_pass = matches[match]["stats"]["total_pass"]
    total_scoring_att = matches[match]["stats"]["total_scoring_att"]
    total_tackle = matches[match]["stats"]["total_tackle"]
    touches = matches[match]["stats"]["touches"]
    was_fouled = matches[match]["stats"]["was_fouled"]
    won_contest = matches[match]["stats"]["won_contest"]
    yellow_card = matches[match]["stats"]["yellow_card"]
    
    player_stats.append({
        "player_id": player_id,
        "last_name": player_lastname,
        "calendar": calendar,
        "number_of_matches": number_of_matches,
        "total_appearances": total_appearances,
        "starter": starter,
        "sub_in": sub_in,
        "sub_out": sub_out,
        "match_date": match_date, 
        "season_day": season_day,
        "match_id": match_id,
        "goals": goals,
        "minutes_played": minutes_played,
        "match_rate": match_rate,
        "sub": sub,
        "accurate_back_zone_pass": accurate_back_zone_pass,
        "accurate_cross_nocorner": accurate_cross_nocorner,
        "accurate_fwd_zone_pass": accurate_fwd_zone_pass,
        "accurate_long_balls": accurate_long_balls,
        "accurate_pass": accurate_pass,
        "att_pen_goal": att_pen_goal,
        "att_pen_miss": att_pen_miss,
        "att_pen_post": att_pen_post,
        "att_pen_target": att_pen_target,
        "big_chance_created": big_chance_created,
        "big_chance_missed": big_chance_missed,
        "clean_sheet": clean_sheet,
        "duel_lost": duel_lost,
        "duel_won": duel_won,
        "error_lead_to_goal": error_lead_to_goal,
        "failed_cross": failed_cross,
        "fouls": fouls,
        "goal_assist": goal_assist,
        "goals": goals,
        "goals_conceded": goals_conceded,
        "interception_won": interception_won,
        "lost_contest": lost_contest,
        "own_goals": own_goals,
        "percentage_pass": percentage_pass,
        "poss_lost_ctrl": poss_lost_ctrl,
        "red_card": red_card,
        "shot_off_target": shot_off_target,
        "total_back_zone_pass": total_back_zone_pass,
        "total_contest": total_contest,
        "total_cross_nocorner": total_cross_nocorner,
        "total_fwd_zone_pass": total_fwd_zone_pass,
        "total_long_balls": total_long_balls,
        "total_pass": total_pass,
        "total_scoring_att": total_scoring_att,
        "total_tackle": total_tackle,
        "touches": touches,
        "was_fouled": was_fouled,
        "won_contest": won_contest,
        "yellow_card": yellow_card
    })
    
    print(f"{match_date} is appended.")
    
```
